# LSTM

```python
# standardization/normalisation
scaler1 = MinMaxScaler(feature_range=(0,1))
scaled_trainX = scaler1.fit_transform(trainX_new.values)
scaled_valX = scaler1.fit_transform(valX_new.values)
scaled_testX = scaler1.fit_transform(testX_new.values)
```

```python
# reshape to 3D
train_X = scaled_trainX.reshape((scaled_trainX.shape[0],1, scaled_trainX.shape[1]))
val_X = scaled_valX.reshape((scaled_valX.shape[0], 1, scaled_valX.shape[1]))
test_X = scaled_testX.reshape((scaled_testX.shape[0]),1, scaled_testX.shape[1])

print(train_X.shape, trainY.shape,
     val_X.shape, valY.shape,
     test_X.shape, testY.shape)
```

```python
myvalfold = []
for i in range(len(train_X)):
    myvalfold.append(-1)
for i in range(len(val_X)):
    myvalfold.append(0)

ps = PredefinedSplit(test_fold = myvalfold)

mergedX = np.concatenate((train_X, val_X), axis=0)
mergedY = np.concatenate((trainY.values, valY.values), axis=0)
```

```python
#mergedX = mergedX[~np.isnan(mergedX)]
#mergedY = mergedY[~np.isnan(mergedY)]
```

```python
mergedX.shape
```

```python
def create_model(lr, beta_1, beta_2, neurons1, neurons2, neurons3, dropout1, dropout2, dropout3):
    model = Sequential()
    model.add(LSTM(neurons1, input_shape = (train_X.shape[1], train_X.shape[2]), return_sequences=True))
    model.add(Dropout(dropout1))
    model.add(LSTM(neurons2, return_sequences=True))
    model.add(Dropout(dropout2))
    model.add(LSTM(neurons3))
    model.add(Dropout(dropout3))
    model.add(Dense(1))
    
    optimizer = Adam(lr=lr, beta_1=beta_1, beta_2=beta_2)
    model.compile(loss='mean_squared_error',optimizer=optimizer)
    return model

model = KerasRegressor(build_fn=create_model, epochs=150, batch_size=10, verbose=0)

params = {
    'lr': [0.001, 0.01, 0.1, 0.2, 0.3],
    'beta_1': [0.0, 0.2, 0.4, 0.6, 0.8, 0.9],
    'beta_2': [0.0, 0.2, 0.4, 0.6, 0.8, 0.9],
    'neurons1': list(range(2, 17, 2)),
    'neurons2': list(range(8, 33, 4)),
    'neurons3': list(range(16, 65, 4)),
    'dropout1':np.linspace(0,0.9,10).tolist(),
    'dropout2':np.linspace(0,0.9,10).tolist(),
    'dropout3':np.linspace(0,0.9,10).tolist(),
    'epochs':[40,60,80,100],
    'batch_size': [10,20,30,40]
}

grid = RandomizedSearchCV(model, params, cv=ps, verbose=2, n_iter=20, scoring='neg_mean_squared_error')
grid.fit(mergedX, mergedY)
```

```python
print(grid.best_params_)
```

```python
# Summarize history for loss
plt.plot(grid.history['loss'])
plt.plot(grid.history['val_loss'])
plt.title('Model Loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['training','validation'], loc='upper right')
plt.show()
```
